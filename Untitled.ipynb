# Testing

```python
% matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['figure.figsize'] = (15, 6)
sns.set_style("darkgrid")
```

```python
import numpy as np 
import GPy

from src.bayesian_linear_regression import BayesianLinearRegression
from src.models import GPyBOModel, BOModel
from src.neural_network import TorchRegressionModel

def f(x):
  return np.sinc(x * 10 - 5).sum(axis=1)[:, None]

rng = np.random.RandomState(42)
# x_train = rng.uniform(0, 1, 10)[:, None]
x_train = rng.uniform(0, 1, 20)
x_train = x_train[(x_train < 0.4) | (0.6 < x_train)]
x_train = x_train[:, None]
y_train = f(x_train)

x = (np.linspace(0, 1, 100))[:,None]
y = f(x)

def gpy():
    kernel = GPy.kern.RBF(1)
    kernel.variance.set_prior(GPy.priors.LogGaussian(0.005, 0.5))
    return GPyBOModel(kernel=kernel, num_mcmc=0, fix_noise=True)

def dngo():
    nn = TorchRegressionModel(dim_basis=50, epochs=1000, batch_size=10)
    reg = BayesianLinearRegression(num_mcmc=6, burn_in=1000, mcmc_steps=1000)
    model = BOModel(nn, regressor=reg)
    return model

def ensemble():
    nn = TorchRegressionModel(dim_basis=50, epochs=1000, batch_size=10)
    reg = BayesianLinearRegression(num_mcmc=6, burn_in=1000, mcmc_steps=1000)
    model = BOModel(nn, regressor=reg, num_nn=5)
    return model

models = [m() for m in [ensemble, dngo, gpy]]
# models = [gpy]

for i, model in enumerate(models):
    plt.subplot(1, 3, i + 1)
    model.init(x_train, y_train)
    model.plot_prediction(x,y)
```

```python
from src.acquisition_functions import EI

m = models[0]
acq = EI(model=m)
m.plot_acq(np.linspace(0,1,100)[:,None], acq.calc)
```

```python
from src.neural_network import NNRegressionModel
import tensorflow as tf 

def f(x):
  return np.sinc(x * 10 - 5).sum(axis=1)[:, None]

rng = np.random.RandomState(42)
x_train = rng.uniform(0, 1, 20)[:, None]
y_train = f(x_train)

sess = tf.Session()
m = NNRegressionModel(epochs=1000, batch_size=1000)
m.fit(sess, x_train, y_train)
```

```python
from matplotlib import pyplot as plt
X_line = np.linspace(0, 1, 100)[:, None]
y_true = f(X_line)
y_line = m.predict(sess, X_line)
plt.plot(X_line, y_line)
plt.plot(X_line, y_true)
plt.scatter(x_train, y_train)
```

```python
m = NNRegressionModel(epochs=1000, batch_size=1000)
sessions = [tf.Session() for _ in range(10)]
predictions = []
for sess in sessions:
    idx = np.arange(x_train.shape[0])
    np.random.shuffle(idx)
    idx = idx[:int(x_train.shape[0] * 1)]
    m_y_train = y_train[idx]
    m_x_train = x_train[idx]

    m.fit(sess, m_x_train, m_y_train)
for sess in sessions:
    y_line = m.predict(sess, X_line)
    predictions.append(y_line)
    plt.plot(X_line, y_line)

    y_pred = np.stack(predictions).max(0)
plt.plot(X_line, y_pred)

plt.plot(X_line, y_true)
plt.scatter(x_train, y_train)
```
